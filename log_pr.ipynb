### ---------------- Shipment Class ----------------

```python
# ---------------- Shipment Class ----------------
class Shipment:
    def __init__(self, tracking_id, sender, receiver, weight):
        self.tracking_id = tracking_id
        self.sender = sender
        self.receiver = receiver
        self.weight = weight
        self.status = "Pending"

    def calculate_cost(self):
        return self.weight * 50

    def update_status(self, new_status):
        self.status = new_status

    def display_details(self):
        print("\nTracking ID:", self.tracking_id)
        print("Sender:", self.sender)
        print("Receiver:", self.receiver)
        print("Weight:", self.weight, "kg")
        print("Status:", self.status)
        print("Cost: ₹", self.calculate_cost())


# ---------------- Logistics System Class ----------------
class LogisticsSystem:
    def __init__(self):
        self.shipments = []

    def create_shipment(self):
        tracking_id = input("Enter Tracking ID: ")
        sender = input("Enter Sender Name: ")
        receiver = input("Enter Receiver Name: ")
        
        try:
            weight = float(input("Enter Weight (kg): "))
        except:
            print("Invalid weight!")
            return

        shipment = Shipment(tracking_id, sender, receiver, weight)
        self.shipments.append(shipment)

        print("Shipment Created Successfully!")

    def view_shipments(self):
        if not self.shipments:
            print("No shipments available.")
        else:
            for shipment in self.shipments:
                shipment.display_details()

    def update_shipment(self):
        tid = input("Enter Tracking ID: ")
        for shipment in self.shipments:
            if shipment.tracking_id == tid:
                new_status = input("Enter New Status (In Transit/Delivered): ")
                shipment.update_status(new_status)
                print("Status Updated Successfully!")
                return
        print("Shipment Not Found!")


# ---------------- Main Program ----------------
def main():
    system = LogisticsSystem()

    while True:
        print("\n===== Logistics Management System =====")
        print("1. Create Shipment")
        print("2. View Shipments")
        print("3. Update Shipment Status")
        print("4. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            system.create_shipment()

        elif choice == "2":
            system.view_shipments()

        elif choice == "3":
            system.update_shipment()

        elif choice == "4":
            print("Thank You!")
            break

        else:
            print("Invalid choice!")


if __name__ == "__main__":
    main()
```
